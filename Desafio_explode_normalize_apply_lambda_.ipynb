```python
import pandas as pd
import json
import requests
```

```python
#Lendo arquivo http json

url3 = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'
response3 = requests.get(url3)
vendas = pd.DataFrame(response3.json())
vendas
```

```python
# Consultando colunas
vendas.columns
```

```python
# Aplicando normalizacao de dados em um dicionario
dados = pd.json_normalize(vendas['dados_vendas'])
dados
```

```python
# Transforma os nomes das colunas em uma lista
colunas = list(dados.columns)
colunas
```

```python
# Metodo Explode para explodir listas dentro dos dicionarios de cada linha da coluna

dados = dados.explode(column= 'Cliente')
dados.reset_index(inplace=True, drop=True)
dados.head()
```

```python
# O metodo explode tem de ser feito individualmente por tabelas de diferentes types

dados = dados.explode(column= 'Valor da compra')
dados.reset_index(inplace=True, drop=True)
dados
```

```python
# Informações das colunas
dados.info()
```

```python
import numpy as np
```

```python
# Corrigindo valores para aplicaçao do astype
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R$ ', '').replace(',','.').strip())

# Alterar os tipo de dado para float
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
```

```python
dados.head(2)
```

```python
#Padronizando a tabela Cliente

dados['Cliente'] = dados['Cliente'].str.lower()
```

```python
# Removendo tudo exeto letras

dados['Cliente'] = dados['Cliente'].str.replace('[^a-z]',' ',regex=True)

dados['Cliente'] = dados['Cliente'].str.split()
```

```python
dados.head()
```

```python
dados.info()
```

```python
# Transformando a coluna data de venda em datetime
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'], format='%d/%m/%Y')
```

```python
# Calculo total de arrecadação em compras por cada cliente

# total_compras = dados.groupby(['Cliente'])['Valor da compra'].sum()

dados['Cliente'] = dados['Cliente'].apply(lambda x: ' '.join(x))
```

```python
dados.head(2)
```

```python
# Agrupando por clientes e somando os valores das compras

total_compras = dados.groupby(['Cliente'])['Valor da compra'].sum()
total_compras.sort_values(ascending=False)
```

```python
url4 = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'
response4 = requests.get(url4)
locacao = pd.DataFrame(response4.json())
locacao.head()
```

```python
# normalizando dados de um dicionario para um dataframe

loc = pd.json_normalize(locacao['dados_locacao'])
loc.head(2)
```

```python
# Explodindo as listas de cada linha do dataframe.
loc = loc.explode(column= 'valor_aluguel')
loc.reset_index(inplace=True, drop=True)
loc.head(2)
```

```python
#Normalizando dados para aplicaçao do astype
loc['valor_aluguel'] = loc['valor_aluguel'].apply(lambda x: x.replace('$ ','').replace(',','.').replace('reais','').strip())
loc.head(2)
```

```python
#transformando tipo do dado
loc['valor_aluguel'] = loc['valor_aluguel'].astype(np.float64)
```

```python
loc.info()
```

```python
loc.head(2)
```

```python
#Removendo (blocoAP) da coluna apartamento

loc['apartamento'] = loc['apartamento'].str.replace('\(blocoAP\)','',regex= True)
```
