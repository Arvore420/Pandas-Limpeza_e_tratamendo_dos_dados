```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/manipulacao strings - banco_clientes.csv')
```

```python
dados.head(2)
```

```python
# str necessario para aplicar em toda coluna / upper - deixa tudo maiusculo
dados['nome'] = dados['nome'].str.upper()
```

```python
dados.head(2)
```

```python
#Removendo parenteses do da coluna telefone

dados['telefone'] = dados['telefone'].str.replace('(','', regex=False).str.replace(')','', regex= False)
```

```python
dados.head(2)
```

```python
# extrair o padrao do cep da coluna endereço -> extract
# \d = uma sequencia de numeros

dados['endereco'].str.extract('(\d{5}-\d{3})')
```

```python
# criando uma nova coluna com cep
dados['cep'] = dados['endereco'].str.extract('(\d{5}-\d{3})')
```

```python
# removendo o CEP da coluna endereço
# strip remove os espaços em branco (preventivo)
dados['endereco'] = dados['endereco'].str.replace('\d{5}-\d{3}','',regex=True).str.strip()
```

```python
dados
```

```python
# metodo split() divide coluna em 2
# split nao funciona corretamente pq temos numeros que nao sao separados por virgulas

dados['endereco'].str.split(',',expand = True)
```

```python
# fazer um split com replace ANTES
# regex dentro split
# \s sequencia de caracteres () buscar um conjunto
# tecnica look head ?= / se tiver algo a frente retorna true
# \d sequencia numnerica
# +$ especifica que é para ser feio na ultima sequencia de numero
# + indica que tem mais de uma sequencia / $ indica que e a ultima sequencia

dados['endereco'].str.split('\s(?=\d+$)',expand = True)
```

```python
# remover as virgulas que ficaram nos nomes das ruas
# replace tem que ser feito ANTES do split
# pq = str é aplicado em um siries (coluna), ja o resultado do split é um dataframe (tabela)
# primeiro remover a virgula(replace), depois aplicar split

dados['endereco'].str.replace(',','').str.split('\s(?=\d+$)',expand = True)
```

```python
# Salvando resultado em 2 novas colunas, rua e numero

dados[['rua','numero']] = dados['endereco'].str.replace(',','').str.split('\s(?=\d+$)', regex=True, expand = True)
```

```python
dados.head()
```

```python
# Remover coluna endereco com DROP

dados = dados.drop(['endereco'], axis=1)
dados
```
