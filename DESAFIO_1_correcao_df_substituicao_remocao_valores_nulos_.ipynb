```python
import pandas as pd
import json
```

```python
# Lendo arquivo Json

with open('/content/limpeza e trat _ DESAFIO_cursos_cadastrados.json') as x:
  dados = json.load(x)
```

```python
dados
```

```python
# Normalizando json

df = pd.json_normalize(dados)
```

```python
df.head()
```

```python
# Verificando se existem valores nulos

df.isna().sum().sum()
```

```python
# Retorna as linhas que possuem pelo menos uma coluna nulas

df[df.isna().any(axis=1)]
```

```python
# Remover todos os valores ausentes com dropna

df.dropna(inplace=True)
```

```python
# Resetando index da tabela

df.reset_index(drop=True, inplace=True)
```

```python
df
```

IDENTIFICAR E REMOVER AMOSTRAS DUPLICADAS

```python
#verificando se ha valores duplicados

df.duplicated().sum()
```

```python
# Removendo valores duplicados

df.drop_duplicates(inplace=True)
```

REMOVER AMOSTRAR COM STRINGS VAZIAS

```python
# Buscar amostra vazia na coluna 'instrutor.nome'

df[df['instrutor.nome']=='']
```

```python
# Buscar amostra vazia na coluna 'data_conclusao'

df[df['data_conclusao']=='']
```

```python
# Sunstituindo strings vazias por valores NAN para ficar mais facil de trabalhar

df.replace('', pd.NA, inplace=True)
```

```python
df
```

```python
#Removendo valores nulos

df.dropna(inplace=True)
```

```python
df
```
